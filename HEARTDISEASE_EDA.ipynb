EDA ON HEART-DISEASE-TRAIN-TEST

```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/HeartDiseaseTrain-Test.csv')
print(df.info())
df.describe()
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
cholesterol_by_sex = df.groupby('sex')['cholestoral'].mean()
print(cholesterol_by_sex)

age_by_disease = df.groupby('target')['age'].mean()
print(age_by_disease)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 3))
sns.histplot(df['age'], bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.boxplot(x='target', y='cholestoral', data=df)
plt.title('Cholesterol Levels by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x='sex', hue='target', data=df)
plt.title('Heart Disease by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
sns.catplot(x='sex', y='cholestoral', hue='target', kind='box', data=df)
plt.title('Cholesterol Levels by Sex and Heart Disease Status')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Cholesterol')
plt.show()
```

```python
heart_disease_patients = df.loc[df['target'] == 1]
print(heart_disease_patients.head())

average_cholesterol_hd = heart_disease_patients['cholestoral'].mean()
print(f'Average Cholesterol Level of Patients with Heart Disease: {average_cholesterol_hd}')
```
