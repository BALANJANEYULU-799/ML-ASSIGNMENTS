1.
Basic Line Plot
o
Plot a simple line graph of ùë¶=ùë•2 for ùë• values ranging from -10 to 10.
o
Label the x-axis as "X Values" and the y-axis as "Squared Values."
o
Add a title to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = x**2
plt.figure(figsize=(5,3))
plt.plot(x, y, label='y = x^2', color='blue')
plt.xlabel('X Values', fontsize=14)
plt.ylabel('Squared Values', fontsize=14)
plt.title('Line Plot of y = x^2', fontsize=16)
plt.grid()
plt.legend()
plt.show()
```

3.
Scatter Plot
o
Generate two lists of 20 random integers each (for x and y coordinates).
o
Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = np.random.randint(1, 100, size=20)
y = np.random.randint(1, 100, size=20)
plt.figure(figsize=(6, 3))
scatter = plt.scatter(x, y, s=100, c='blue', alpha=0.6, edgecolors='w', linewidth=2)

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=9, ha='right')
plt.xlabel('X Coordinates', fontsize=14)
plt.ylabel('Y Coordinates', fontsize=14)
plt.title('Scatter Plot of Random Points', fontsize=16)
plt.grid()
plt.show()
```

4.
Histogram
o
Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.
o
Customize the number of bins and add a grid to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(100)
plt.figure(figsize=(5, 3))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Random Values from Normal Distribution', fontsize=16)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

5.
Pie Chart
o
Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].
o
Display the percentage values on the chart and use a color palette for differentiation.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.figure(figsize=(3, 3))
colors = plt.cm.Paired(range(len(categories)))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Percentage Distribution of Categories', fontsize=16)
plt.axis('equal')
plt.show()
```

7.
Heatmap (Seaborn)
o
Create a 5x5 matrix with random integers between 1 and 10 using np.random.randint() and visualize it as a heatmap using Seaborn.
o
Add a color bar and annotate the heatmap with the values.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
matrix = np.random.randint(1, 11, size=(5, 5))
plt.figure(figsize=(5, 3))
sns.heatmap(matrix, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.title('Heatmap of Random 5x5 Matrix', fontsize=16)
plt.xlabel('Columns', fontsize=14)
plt.ylabel('Rows', fontsize=14)
plt.show()
```

8.
Comparison of Matplotlib and Seaborn
o
Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn.
o
Highlight the differences in the code, syntax, and aesthetics of the two libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]

plt.figure(figsize=(6, 3))

plt.subplot(1, 2, 1)
plt.bar(students, marks, color='skyblue', edgecolor='black')
plt.title('Matplotlib Bar Chart', fontsize=16)
plt.xlabel('Students', fontsize=14)
plt.ylabel('Marks', fontsize=14)
plt.grid(axis='y', alpha=0.7)

plt.subplot(1, 2, 2)
sns.barplot(x=students, y=marks, color='skyblue', edgecolor='black')
plt.title('Seaborn Bar Chart', fontsize=16)
plt.xlabel('Students', fontsize=14)
plt.ylabel('Marks', fontsize=14)

plt.tight_layout()
plt.show()
```

Pair Plot (Seaborn)
o
Manually create a small dataset with columns such as 'Age', 'Height', 'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Age': [23, 25, 31, 22, 28, 35, 30, 27, 29, 24],
    'Height': [165, 170, 180, 160, 175, 185, 178, 172, 169, 167],
    'Weight': [55, 70, 80, 50, 65, 90, 75, 68, 72, 60]
}

df = pd.DataFrame(data)
sns.pairplot(df)
plt.suptitle('Pair Plot of Age, Height, and Weight', y=1.02, fontsize=16)
plt.show()
```

10.
Save and Style Plots
‚Ä¢
Plot any graph of your choice and save it in both .png and .pdf formats.
‚Ä¢
Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain how the style impacts the visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.style.use('ggplot')
plt.figure(figsize=(5, 3))
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2)


plt.xlabel('X Values (radians)', fontsize=14)
plt.ylabel('Sine Values', fontsize=14)
plt.title('Sine Wave Plot', fontsize=16)
plt.legend()

plt.savefig('sine_wave_plot.png', format='png', dpi=300)
plt.savefig('sine_wave_plot.pdf', format='pdf')

plt.grid()
plt.show()
```
